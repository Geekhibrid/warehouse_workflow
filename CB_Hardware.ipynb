### Fill in missing values using the Fillna method

```python
cb_hardware.fillna({
    'Opportunity_id': 0,
    'sales_agent': 'unknown',
    'product': 'unknown',
    'deal_stage': 'unknown',
    'close_value': 0,
    'manager': 'unknown',
    'regional_office': 'unknown',
    'series': 'unknown',
    'sales_price': 0,
    'account': 'unkown',
    'employees': 0
}, inplace=True)
```

```python
cb_hardware
## We made sure that the string values were labeled, "unknown" and the ints as "0"
```

```python
cb_hardware.drop(columns=['employees', 'sales_price'], inplace = True)
```

```python
cb_hardware

## Now we're going to drop two more columns, employees and sales_price, to narrow our focus even more. 
```

```python
cb_hardware = cb_hardware[sorted(cb_hardware.columns)]
cb_hardware

## We wanted to sort the columns to see the actual values of the chsen columns
```
